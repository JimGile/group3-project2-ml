```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
import os
import sys
```

```python
file_paths = ['/Users/timwillard/Desktop/group3-project2-ml/Data/Clean/combined/Atlanta_combined.csv',
              '/Users/timwillard/Desktop/group3-project2-ml/Data/Clean/combined/Charlotte_combined.csv',
              '/Users/timwillard/Desktop/group3-project2-ml/Data/Clean/combined/Cincinnati_combined.csv',
              '/Users/timwillard/Desktop/group3-project2-ml/Data/Clean/combined/Denver_combined.csv',
              '/Users/timwillard/Desktop/group3-project2-ml/Data/Clean/combined/Indianapolis_combined.csv',
              '/Users/timwillard/Desktop/group3-project2-ml/Data/Clean/combined/Providence_combined.csv',
              '/Users/timwillard/Desktop/group3-project2-ml/Data/Clean/combined/Tampa_combined.csv'
              ]
```

```python
def rename_columns_for_prophet(data):
    # Attempt to infer the date and value columns
    possible_date_columns = ['DATE', 'ds']
    possible_value_columns = ['MeanSalePrice', 'y']
    
    # Find the column for 'ds' (date)
    date_column = None
    for col in possible_date_columns:
        if col in data.columns:
            date_column = col
            break

    # Find the column for 'y' (value)
    value_column = None
    for col in possible_value_columns:
        if col in data.columns:
            value_column = col
            break

    if not date_column or not value_column:
        raise ValueError("Couldn't find appropriate columns for date and value")

    # Rename columns to 'ds' for date and 'y' for value as required by Prophet
    renamed_data = data.rename(columns={date_column: 'ds', value_column: 'y'})
    
    return renamed_data
```

```python
# Loop through each file, load the CSV, and apply Prophet
for file_path in file_paths:
    try:
        print(f"\nProcessing file: {os.path.basename(file_path)}")  # Log the file being processed
        sys.stdout.flush()  # Ensure output is printed immediately

        # Load the dataset
        data = pd.read_csv(file_path)

        # Dynamically rename columns for Prophet
        df = rename_columns_for_prophet(data)

        # Initialize the Prophet model
        model = Prophet()

        # Fit the model
        model.fit(df)

        # Make a future dataframe for 12 periods (e.g., months if your data is monthly)
        future = model.make_future_dataframe(periods=12, freq='M')  # Adjust 'M' if your data frequency is not monthly

        # Generate the forecast
        forecast = model.predict(future)

        # Print the forecast
        print(f"\n--- Forecast for {os.path.basename(file_path)} ---")
        print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(12))
        sys.stdout.flush()  # Ensure the forecast is printed immediately

        # Plot the forecast
        fig = model.plot(forecast)
        plt.title(f"Forecast for {os.path.basename(file_path)}")
        plt.show()
        plt.close()

        # Save the forecast to a CSV
        forecast_output = f"{os.path.basename(file_path).replace('.csv', '')}_prophet_forecast.csv"
        forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].to_csv(forecast_output, index=False)

        print(f"Forecast saved to {forecast_output}\n")
        sys.stdout.flush()

    except Exception as e:
        print(f"Failed to process {file_path}: {e}")
        sys.stdout.flush()
```
