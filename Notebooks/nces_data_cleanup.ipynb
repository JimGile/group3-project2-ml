```python
import pandas as pd
```

```python
# Load data
base_write_path = '../Data/NCES/clean/'
file_path = '../Data/NCES/EmployeePopulation_tabn104.95.xlsx'
emp_pop_df = pd.read_excel(file_path)
emp_pop_df.head()
```

```python
# Limit our scope of cities to Denver, Charlotte, Cincinnati, Providence, Tampa, Atlanta, and Indianapolis
city_names = ['Denver', 'Charlotte', 'Cincinnati', 'Providence', 'Tampa', 'Atlanta', 'Indianapolis']
selected_emp_pop_df = emp_pop_df.copy().loc[emp_pop_df['CityName'].isin(city_names)]
selected_emp_pop_df['CityName'] = selected_emp_pop_df['CityName'].apply(lambda x: x.upper())
selected_emp_pop_df
```

```python
selected_emp_pop_df.to_csv(f'{base_write_path}employee_population_by_city.csv', index=False)
```
