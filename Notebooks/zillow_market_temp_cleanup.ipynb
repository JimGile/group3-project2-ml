```python
import pandas as pd
```

```python
path = '../Data/Zillow/Metro_market_temp_index_uc_sfrcondo_month.csv'
df = pd.read_csv(path)
df.head()
```

```python
df = df.copy()[1:]
```

```python
df['CityName'] = df['RegionName'].str.split(',').str[0]
df.head()
```

```python
cols_to_drop = ['RegionName', 'SizeRank', 'RegionType']
df = df.drop(columns=cols_to_drop)
df.head()
```

```python
column_mapping = {col: col[:7] for col in df.columns if col.find('-') > 0}
df = df.rename(columns=column_mapping)
df
```

```python
new_col_order = ['RegionID', 'CityName', 'StateName'] + list(column_mapping.values())
new_col_order[:5]
```

```python
df_final = df[new_col_order]
df_final
```

```python
df_final['CityName'] = df_final['CityName'].apply(lambda x: x.upper())
df_final.head()
```

```python
path = '../Data/Zillow/clean/city_market_temp_index_month.csv'
df_final.to_csv(path, index=False)
```
